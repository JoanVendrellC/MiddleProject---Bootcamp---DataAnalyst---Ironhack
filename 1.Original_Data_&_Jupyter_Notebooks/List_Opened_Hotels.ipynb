```python
def cleanCategoria(x):
    if x in ['GL']:
        return '5'
    elif x in ['4*S']:
        return '4'
    elif x in ['4*']:
        return '4'
    elif x in ['3*']:
        return '3'
    elif x in ['2*']:
        return '2'
    elif x in ['1*']:
        return '1'
    elif x in ['5*']:
        return '5'
    else:
        return x
Open_Hotels['Categoria'] = list(map(cleanCategoria, Open_Hotels['Categoria']))
```
